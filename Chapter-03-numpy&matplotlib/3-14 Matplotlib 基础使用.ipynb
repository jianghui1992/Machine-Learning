## 绘制plot曲线

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 生成100个点 x
x = np.linspace(-10, 10, 100)
y = x * x
# 绘图
plt.plot(x, y)
# 展示
plt.show()
```

```python
x = np.linspace(0,10,100)

y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1,color='green')
plt.plot(x, y2)
plt.show()
```

```python
# linestyle 指定线条样式
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
# xlim/ylim 指定坐标轴的范围
plt.xlim(-5, 15)
plt.ylim(-2, 2)
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
# 效果同上
plt.axis([-5,15,-2,2])
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':')
plt.plot(x, y2)
# 给轴起名字
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

```python
# label指定线条名字
plt.plot(x, y1, color='green', linestyle=':',label='sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.xlabel("x axis")
plt.ylabel("y axis")
# 显示线条名字
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, color='green', linestyle=':',label='sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
# 设置图表名字
plt.title('This is my first plot')
plt.show()
```

## 绘制scatter散点图

```python
plt.scatter(x, y1)
plt.show()
```

```python
plt.scatter(x, y1, color='green')
plt.scatter(x, y2)
plt.show()
```

```python
# normal 均值为0,标准差为1,随机去10000个数
x = np.random.normal(0,1,10000)
y = np.random.normal(0,1,10000)
# alpha 设置透明度 0:完全透明, 1: 完全不透明
plt. scatter(x,y,alpha=0.1)
plt.show()
```
